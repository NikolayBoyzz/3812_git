```python
penguins_copy = penguins.copy()
height_group_Conditions = [
    (penguins_copy['bill_length_mm'] <  35),
    (penguins_copy['bill_length_mm'] >= 35) & (penguins_copy['bill_length_mm'] < 42),
    (penguins_copy['bill_length_mm'] >= 42)]

height_group_Categories = ['low', 'middle', 'high']

penguins_copy.insert(3, 'height_group', np.NaN)

penguins_copy['height_group'] = np.select(height_group_Conditions, height_group_Categories)

penguins_copy.head(20)
```

```python
sns.histplot(data=penguins_copy, x="flipper_length_mm", hue="height_group")
```
